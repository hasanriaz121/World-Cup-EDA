<h4 style="color:skyblue">Data Pre-Processing

<h5><li>Remove redundant columns

```python
nan_columns=matches.columns[matches.isna().all()]
nan_columns
```

```python
nan_columns=nan_columns[2:]
matches=matches.drop(columns=nan_columns)
matches.head()
```

```python
matches.isna().sum()
```

<h5><li>Get Team Names and Toss decision

```python
match_innings=innings.merge(matches[['id','team1Id','team2Id','matchType','toss.elected','matchDate']],left_on='matchId',right_on='id',suffixes=('_innings', '_matches'))
# merge(innings, matches, on='matchid', suffixes=('_innings', '_matches'))
```

```python
print(innings.shape)
print(match_innings.shape)
```

```python
match_innings.head()
```

```python
def getTeamIds(a):
    if(a.inningsNumber==1):
        a.battingTeamId=a.team1Id
        a.bowlingTeamId=a.team2Id
    else:
        a.battingTeamId=a.team2Id
        a.bowlingTeamId=a.team1Id
    
    return a
```
